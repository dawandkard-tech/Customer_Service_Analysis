```python
## now there is a 37 null values in rating when we will replace null we will replace it with mean of particular catogargy because every catogaory has different rating criteria
##for this i will use transform
df1['Review Rating']=df1.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.mean()))
```

## create column age_group to see which age group people contribute more in sales

```python
label=['young_adult', 'adult', 'middle_aged','senior']
df1['age_group']=pd.qcut(df1['age'],q=4,labels=label)
```

```python
df1[['age','age_group']].head(10)
```
